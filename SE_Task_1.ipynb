```python
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def read_coefficients_from_file(filename):
    with open(filename, 'r') as f:
        coefficients = [float(line.strip()) for line in f]
    return coefficients

def main():
    time_values = np.linspace(0, 10, 50)

    try:
        coefficients = read_coefficients_from_file('coefficients.txt')
        if len(coefficients) != 3:
            raise ValueError("The coefficients file should contain exactly three lines.")
        a, b, c = coefficients
    except FileNotFoundError:
        print("Error: coefficients.txt not found.")
        return
    except ValueError as e:
        print(f"Error: {e}")
        return

    temperature_from_file = quadratic_model(time_values, a, b, c)

    plt.plot(time_values, temperature_from_file, label=f'Coefficients from File\n a = {a}, b = {b}, c = {c}')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Coefficients from File)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    coeffs = [1.5, -2, 5]

    a, b, c = coeffs
    temperature_values = quadratic_model(time_values, a, b, c)

    label = f'Coefficients Set\n a = {a}, b = {b}, c = {c}'

    plt.plot(time_values, temperature_values, label=label)

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation - Single Input')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    parameters_sets = [
        {'a': 1.0, 'b': 2.0, 'c': 3.0},
        {'a': 0.5, 'b': 1.0, 'c': 2.0},
        {'a': 2.0, 'b': -1.0, 'c': 5.0}
    ]

    for params in parameters_sets:
        a = params['a']
        b = params['b']
        c = params['c']

        temperature_calculated = quadratic_model(time_values, a, b, c)

        plt.plot(time_values, temperature_calculated, label=f'a={a}, b={b}, c={c}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Multiple Lines)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    a_hardcoded = 1
    b_hardcoded = -2
    c_hardcoded = 5

    a_user = float(input("Enter the value of a: "))
    b_user = float(input("Enter the value of b: "))
    c_user = float(input("Enter the value of c: "))

    temperature_hardcoded = quadratic_model(time_values, a_hardcoded, b_hardcoded, c_hardcoded)
    temperature_user = quadratic_model(time_values, a_user, b_user, c_user)

    plt.plot(time_values, temperature_hardcoded, label='Hardcoded Coefficients: a=6,b=2,c=3')
    plt.plot(time_values, temperature_user, label='User Input Coefficients a=5,b=2,c=3', linestyle='--')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    with open('parameters.txt', 'r') as file:
        for line in file:
            a, b, c = map(float, line.strip().split())

            temperature_calculated = quadratic_model(time_values, a, b, c)

            plt.plot(time_values, temperature_calculated, label=f'a={a}, b={b}, c={c}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation Multiple (File Input)')
    plt.show()

if __name__ == "__main__":
    main()
```
